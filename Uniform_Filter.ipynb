**Blur Image**

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

img = cv2.imread('/content/download.jpg')
blur_filter = np.array([
    [1, 1, 1],
    [1, 1, 1],
    [1, 1, 1]
])

blur_filter = blur_filter / 9
blur_img = cv2.filter2D(img, ddepth=-1, kernel=blur_filter)
cv2_imshow(blur_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

**No Filter / Brightness**

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

img = cv2.imread('/content/download.jpg')
no_filter = np.array([
    [0, 0, 0],
    [0, 1, 0],
    [0, 0, 0]
])

no_filter = no_filter * 2
bright_img = cv2.filter2D(img, ddepth=-1, kernel=no_filter)
cv2_imshow(bright_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
b_img = cv2.blur(img, ksize=(11,11))
cv2_imshow(b_img)
```

**Gaussian Blur**

```python
g_img = cv2.GaussianBlur(img, ksize=(11, 11), sigmaX=30, sigmaY=30)
cv2_imshow(g_img)
```

**Sharpen**

```python
s_filter = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

# s_filter = s_filter / 9
s_img = cv2.filter2D(img, ddepth=-1, kernel=s_filter)
cv2_imshow(s_img)
```

**Edge Detection / Laplacian**

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_img = cv2.GaussianBlur(img, ksize=(3, 3), sigmaX=1, sigmaY=1)
edges = cv2.Laplacian(gray_img, ddepth=-1)
cv2_imshow(edges)
```
